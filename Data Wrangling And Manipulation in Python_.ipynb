###  Data Wrangling and Manipulation with Python

##### Why do we need NumPy library, let's consider the following

- Numpy is one of the fundamental packages for scientific computing in Python.

###### Numpy Arrays
- Or NdArrays (n-dimensional array)
- They are like lists in Python however they allow faster computation
    1. They are stored as one contiguous block of memory, rather than being spread out across multiple locations like a list. 
    2. Each item in a numpy array is of the same data type (i.e. all integers, all floats, etc.), rather than a conglomerate of any number of data types (as a list is). We call this idea homogeneity, as opposed to the possible heterogeneity of Python lists.

- The python package responsible for performing analysis is called pandas (yes like kunfu *panda* #_#)
- Pandas is a set of high level data structures and manipulation tool that makes data analysis with python quick & simple

###### Series
Series is a 1-dimensional array-like object containing an array of data, consider the following example, Let's start by initializing our numpy and pandas package and initializing it.

#### Creating a Series

```python
# Creating an  array
label = ['i','ii','ii']
s_data = [10,20,30]
arr = np.array(s_data)
```

```python
# Creating series from numerical data
series_data = pd.Series(data=s_data)
print(series_data)
```

```python
# Creating series from numerical data and corresponding index
iseries_data = pd.Series(data=s_data, index=label)
print(iseries_data)
```

```python
# Creating series with dictionary
dict = {'a':10,'b':20,'c':30}
dict_series = pd.Series(dict)
print(dict_series)
```

##### Checking for missing values and filling

```python
# Let's create a dataframe since our data does not have any missing values
data = {'Name': ['Jai', 'Princi', 'Gaurav',
                 'Anuj', 'Ravi', 'Natasha', 'Riya', 'Jai', 'Princi'],
        'Age': [17, 17, np.nan, 17, 18, np.nan, 17, 17, 17],
        'Gender': ['M', 'F', 'M', 'M', 'M', 'F', 'F', 'M', 'F'],
        'Marks': [90, 76, np.nan, 74, 65, np.nan, 71, 90, 76]}
 
# Convert into DataFrame
new_df = pd.DataFrame(data)
new_df
```

```python
missing = new_df.isnull().sum()
missing
```

For missing values we can fill our data in several ways: we can fill with mean, median, mode in case of categorical data, forwardfill, backward fill or even with specific numbers or objects. In some cases missing values are dropped.


```python
new_df.info()
```

```python
# Use mean to fill in missing values
mean = new_df.mean()
print(mean)
new_df.fillna(mean, inplace=True)

# Confirm the correction
print(new_df.info())
```

```python
new_df.head(9)
```

```python
# Checking for duplicates in the data
new_df.duplicated().sum()
```

```python
# Drop duplicates
new_df.drop_duplicates(inplace=True)
```

```python
# Confirming the correction
new_df.duplicated().sum()
```
